### DataFrame

- 드디어 엑셀 처럼 생긴 자료형태이다
- 각 열은 다른 데이터형태를 가질 수 있다. (각각 다른 series 이니까..)
- 한 칼럼씩 내부적으로 보면 series 를 모아서 만들어져 있다.

#### DataFrame 생성

```python
raw_data = {'col0': [1, 2, 3, 4],
            'col1': [10, 20, 30, 40],
            'col2': [100, 200, 300, 400]}

df = pd.DataFrame(raw_data)
#print 	
#	col0 col1 col2
#0	1	 10	  100
#1	2	 20	  200
#2	3	 30	  300
#3	4	 40	  400
```

### df.apply() 함수
