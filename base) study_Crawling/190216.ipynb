```python
#멀티 프로세싱
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import time
from multiprocessing import Pool , cpu_count
num_cores = cpu_count()
num_cores
```

```python
chrome_driver = webdriver.Chrome('./datas/chromedriver_windows.exe')
chrome_driver.implicitly_wait(3)
chrome_driver.get("https://www1.president.go.kr/petitions?page=1")
```

```python
html = chrome_driver.page_source #html 소스 확인
html
```

```python
#청와대 국민청원 page 1~10 크롤링
url_start = 'https://www1.president.go.kr/petitions?page=1'
page_num = 1
petition = {}
for i in range(1, 11):
    url = url_start[0:len(url_start)-1] + str(page_num)
    print(url)
    chrome_driver.get(url)
    html = chrome_driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    petition_num = soup.select('body > #wrap > #contents > #cont_view > div.cs_area > div.cs_body > div.wrap > div.board.text > div.b_list.category > div.bl_body > ul > li > div.bl_wrap > div.bl_no') 
    petition_title = soup.select('body > #wrap > #contents > #cont_view > div.cs_area > div.cs_body > div.wrap > div.board.text > div.b_list.category > div.bl_body > ul > li > div.bl_wrap > div.bl_subject') 
    
    for n,m in zip(petition_num, petition_title) :
        key = n.text.strip()
        value = m.text.strip()
        petition[key] = value 
    page_num += 1

chrome_driver.quit() #드라이브 멈추기
```
